```python
#Slide a window size 3 forward on the nums 


nums = [2,1,5,1,3,2]
k = 3 


max_sum = 0
for i in range(len(nums)-k+1):
    current_max = sum(nums[i:i+k])
    if current_max >= max_sum: 
        max_sum = current_max
max_sum
```

```python
def max_subarray_sum(nums, k):
    # Step 1: sum of first window
    window_sum = sum(nums[:k])
    max_sum = window_sum
    
    # Step 2: slide the window
    for i in range(k, len(nums)):
        window_sum += nums[i] - nums[i-k]  # add new, remove old
        max_sum = max(max_sum, window_sum)
    
    return max_sum

print(max_subarray_sum([2, 1, 5, 1, 3, 2], 3))  # Output: 9
```
