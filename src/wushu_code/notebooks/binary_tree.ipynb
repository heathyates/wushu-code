# Binary Tree 

## Definition 
A binary tree is a data structure that is constructured from balanced node. A balanced node is a single root value with two children. If the children do not have any other children themselves, they are called leaves

## Example 

Here is an example of a binary tree

        1
       / \
      2   3
     / \
    4   5



## 

```python
# Definiition for a binary tree node that you may find on leetcode
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val        # value stored in the node
        self.left = left      # left child (another TreeNode or None)
        self.right = right    # right child (another TreeNode or None)


root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

root
```
